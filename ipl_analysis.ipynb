```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the IPL 2024 dataset
df = pd.read_csv('C:/Users/prati/OneDrive/Documents/Project_1/ipl2024.csv')
```

```python
# Basic information about the dataset
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
## 4 columns have 3 missing values each
```

```python
# Fill missing values(want to keep all rows)
```

```python
df['decision'] = df['decision'].fillna('unknown')
df['player_of_the_match'] = df['player_of_the_match'].fillna('N/A')
df['most_runs'] = df['most_runs'].fillna('N/A')
df['most_wkts'] = df['most_wkts'].fillna('N/A')
```

```python
df.isnull().sum()
```

```python
# Now Analyzing most wins by teams
```

```python
# Count how many times each team appears in the 'winner' column
wins = df['winner'].value_counts()
```

```python
# Display the results
print(wins)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x=wins.index, y=wins.values, palette='viridis')
plt.title('Most Wins by Team - IPL 2024', fontsize=16)
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Score Comparison
# First compare the average and plot the distribution of 'First_score' and 'Second_score'
```

```python
# Calculate average scores
avg_first = df['first_score'].mean()
avg_second = df['second_score'].mean()

print(f"Average First Innings Score: {avg_first:.2f}")
print(f"Average Second Innings Score: {avg_second:.2f}")
```

```python
# Plot the distributions
plt.figure(figsize=(10, 6))
sns.kdeplot(df['first_score'], label='First Innings', shade=True, color='blue')
sns.kdeplot(df['second_score'], label='Second Innings', shade=True, color='orange')
plt.axvline(avg_first, color='blue', linestyle='--', label='Avg First')
plt.axvline(avg_second, color='orange', linestyle='--', label='Avg Second')
plt.title('Score Distribution: First vs Second Innings - IPL 2024', fontsize=16)
plt.xlabel('Runs Scored')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Toss Decision Vs Match Outcome
```

```python
# Check how often the toss winner also won the match
same_winner = df[df['toss_winner'] == df['winner']]
toss_win_rate = (len(same_winner)/ len(df)) * 100

print(f"Toss winner also won the match in {toss_win_rate:.2f}% of the matches.")
```

```python
# Group by toss decision and calculate match wins
decision_outcome = df[df['toss_winner'] == df['winner']].groupby('decision').size()
```

```python
# Plot diagram
plt.figure(figsize = (6, 4))
sns.barplot(x = decision_outcome.index, y = decision_outcome.values, palette = 'Set2')
plt.title('Matches Won After Toss by Decision - IPL 2024')
plt.xlabel('Toss Decision')
plt.ylabel('Matches Won')
plt.tight_layout()
plt.show()
```
